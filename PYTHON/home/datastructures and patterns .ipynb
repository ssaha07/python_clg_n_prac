```python
def palindrome(n):
    s=0
    m=n
    while(n>0):
        dig=n%10
        s=s*10+dig
        n=n//10
    if(m==s):
        return m
n=int(input("Enter n: "))
l=[]
for i in range(n):
    x=int(input("Enter no.: "))
    l.append(x)
m=list(filter(palindrome, l))  
print(f"palindrome numbers= {m}")
```

```python
def factorial(n):
    f=1
    for i in range(1, n+1):
        f=f*i
    return f
n=int(input("Enter n:"))
s=0.0
for i in range(n):
    s=s+(1/factorial(i))
print(f"{s:<.2f}")
```

```python
string=input("Enter string: ")
print("----------------------")
print("1. no. of vowels")
print("2. no. of consonents")
print("3. no. of digits")
print("4. no. of special chars")
print("5. no. of whitespaces")
print("6. no. of uppercases")
print("----------------------")

nv=0
nc=0
nd=0
nsp=0
nwsp=0
nup=0
l=[]
while(1):
    ch=int(input("Enter choice"))
    match ch:
        case 1:
            for i in string:
                if i.isalpha():
                    v=i.lower()
                    if v in ('aeiou') and v not in l:
                        l.append(v)
            print(f"No. of vowels = {len(l)}")
        case 2:
            for i in string:
                if i.isalpha():
                    if i not in ('AEIOUaeiou'):
                        nc=nc+1
            print(f"No. of consonants = {nc}")
        case 3:
            for i in string:
                if i.isdigit():
                    nd=nd+1
            print(f"No. of digits = {nd}")
        case 4:
            for i in string:
                if (not i.isalnum() and not i.isspace()):
                    nsp=nsp+1
            print(f"No. of special characters = {nsp}")
        case 5:
            for i in string:
                if i.isspace():
                    nwsp=nwsp+1
            print(f"No. of whitespaces = {nwsp}")
        case 6:
            for i in string:
                if i.isupper():
                    nup=nup+1
            print(f"No. of uppercase = {nup}")
        case _:
            print("Wrong choice")
    x=input("do you want to continue? Y/N")
    if x in 'Nn':
        break
```

```python
def palindrome(m):
    n=m
    s=0
    while(n>0):
        dig=n%10
        s=s*10+dig
        n=n//10
    return m==s
num=int(input("Enter number: "))
if palindrome(num):
    print(f"{num} is palindrome")
else:
    print(f"{num} is not palindrome")
```

```python
def armstrong(m):
    length=len(m)
    n=int(m)
    p=n
    s=0
    while(p>0):
        dig=p%10
        s=s+dig**length
        p=p//10
    return n==s
num=input("enter number: ")
if armstrong(num):
    print(f"{num} is armstrong")
else:
    print(f"{num} is not armstrong")
```

```python
def hcf(a, b):
    n=a
    m=b
    while(m>0):
        n, m = m, n%m
    return n
def lcm(a, b):
    return (a*b)//hcf(a,b)
print(hcf(2, 4))
print(lcm(2, 4))
```
