```python
import numpy as np
myArr=np.array([1,2,'hello',4])
print(myArr, type(myArr))
print(myArr.dtype)
```

```python
arr=np.array([3,6, 32, 7], np.int32)   #fixing the datatype of array
print(arr, type(arr), arr.dtype)
print(arr.shape)          #rowxcoln , length
print(arr.size)
```

```python
arr=np.array([
    [3,4,5,6],
    [1,2,3,4]
])
print(arr, type(arr), arr.dtype)
print(arr.shape)
```

```python
#array creation: conversion from other python ds like list
l=[
    [1, 'harry', 3],
    [5, 85, 7],
    [9, 67.2, 125.578]
]
listArr=np.array(l)      #converting lists to array using typecasting
print(listArr, type(listArr), listArr.dtype)
print(listArr.shape[0])
print(listArr.size)       #number of elements in array
```

```python
zeroarr=np.zeros((2, 5))
print(zeroarr, type(zeroarr), zeroarr.dtype)
```

```python
rng=np.arange(15)
print(rng, rng.shape)
reshape=rng.reshape(3,5)
print(reshape)
```

```python
lspace=np.linspace(1,5, 10) #10 equally spaced nos b2m 1 to 5
print(lspace, type(lspace), lspace.shape, lspace.dtype)
```

```python
empty=np.empty((4,6))
print(empty, type(empty), empty.dtype)
```

```python
empty=np.empty_like(lspace)
print(empty)
```

```python
ide=np.identity(5)
print(ide, ide.size, ide.shape)
```

```python
x=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
arr=np.array(x)
print(arr, arr.shape)
print("Row elements addition: ", arr.sum(axis=0))
print("Columns elements addition: ", arr.sum(axis=1))
print("Transpose addition: \n", arr.T)
print(f"Sum of only 1 row: {arr[0].sum(axis=0)}")
#print(f"Sum of only 1 column: {arr[0].sum()}")
print(f"Trace: {np.trace(arr)}")
print(f"Determinant: {np.linalg.det(arr)}")
```

```python
arr_flat=arr.flatten()
print(f"The flattened array is {arr_flat}")
```
