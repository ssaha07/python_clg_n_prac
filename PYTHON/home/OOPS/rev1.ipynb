**Practical Programming Question (OOP – Python)**

A geometry learning application wants to calculate different types of distances between points and lines in a 2-D coordinate system. Using Object-Oriented Programming in Python, design suitable classes and methods to perform the required calculations.

### Requirements

1. **Create a class `Point`** with the following attributes:

   * `x` : x-coordinate
   * `y` : y-coordinate

2. Implement the following methods in the `Point` class:

   * `distance_from_origin()`
     Returns the distance of the point from the origin (0,0).

   * `distance_from_point(other_point)`
     Returns the distance between the current point and another point.

3. **Create another class `Line`** to represent a line in the form:

[
ax + by + c = 0
]

Attributes:

* `a`
* `b`
* `c`

4. Implement the following methods in the `Line` class:

   * `distance_from_point(point)`
     Returns the perpendicular distance of a given point from the line.

   * `distance_from_line(other_line)`
     Returns the distance between two **parallel lines**.

---

### Mathematical Formulas

Distance between two points:

[
d = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}
]

Distance of a point from origin:

[
d = \sqrt{x^2 + y^2}
]

Distance of a point from a line:

[
d = \frac{|ax + by + c|}{\sqrt{a^2 + b^2}}
]

Distance between two parallel lines:

[
d = \frac{|c_1 - c_2|}{\sqrt{a^2 + b^2}}
]

---

### Driver Program

Write a driver program to:

1. Create two points

   * (P_1(3,4))
   * (P_2(7,1))

2. Create two lines

   * (L_1: 2x + 3y - 6 = 0)
   * (L_2: 2x + 3y + 9 = 0)

3. Display the following results:

   * Distance of (P_1) from origin
   * Distance between (P_1) and (P_2)
   * Distance of (P_1) from line (L_1)
   * Distance between line (L_1) and line (L_2)

---

### Expected Output (Sample)

Distance of P1 from Origin : 5
Distance between P1 and P2 : 5
Distance of P1 from Line L1 : 3.33
Distance between L1 and L2 : 4.16


```python
import math
class Point:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"({self.x},{self.y})"
    def __abs__(self):
        return math.sqrt(self.x**2 + self.y**2)
    def distance_from_point(self, other_point=None):
        if other_point==None:
            return abs(self)
        else:
            sq=math.sqrt((self.x-other_point.x)**2 + (self.y-other_point.y)**2)
            return sq
        
class Line():
    def __init__(self, a, b, c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"{self.a}x + {self.b}y + {self.c} = 0"
    def distance_from_line(self, otherline):
        return (abs(self.c - otherline.c))/math.sqrt(self.a**2 + self.b**2)
    def distance_from_pointline(self, point):
        return (abs((self.a*point.x) + (self.b*point.y) + self.c))/math.sqrt(self.a**2 + self.b**2)
point=Point(3,4)
otherpoint=Point(7,1)
print(f"Point P1:{point}")
print(f"Point P2:{otherpoint}")
print(f"Distance of P1 from origin(0,0): {point.distance_from_point()}")
print(f"Distance of P1 from P2: {point.distance_from_point(otherpoint):<.2f}")
line=Line(2,3,-6)
otherline=Line(2,3,9)
print(f"Line L1: {line}")
print(f"Line L2: {otherline}")
print(f"Distance of L1 from L2: {line.distance_from_line(otherline):<.2f}")
print(f"Distance of L1 from P1: {line.distance_from_pointline(point):<.2f}")
```

- Implement the operations using **class methods only**.

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real=real
        self.imag=imag
    def __str__(self):
        return f"{self.imag}i + {self.real}"
    def add_complex(self, other):
        return f"{self.imag+other.imag}i + {self.real+other.real}"
    def multiply_complex(self, other):
        return f"{self.real*other.imag + self.imag*other.real}i + {self.real*other.real - self.imag*other.imag}"
num1=ComplexNumber(5, 6)
num2=ComplexNumber(4, 8)
print(f"1st complex number: {num1}")
print(f"2nd complex number: {num2}")
print(f"Adding 1st n 2nd complex number: {num1.add_complex(num2)}")
print(f"Multiplying 1st n 2nd complex number: {num1.multiply_complex(num2)}")
```

## Problem Statement

A mathematics learning application needs a **Fraction Calculator** that performs arithmetic operations on fractions.  
Using **Object-Oriented Programming (OOP) in Python**, design a class that represents fractions and performs operations such as **addition, subtraction, multiplication, and division**.

All **inputs and outputs must be in fractional form**.

---

## Requirements

### 1. Create a class `Fraction`

The class should contain the following attributes:

- `numerator` : numerator of the fraction  
- `denominator` : denominator of the fraction  

---

### 2. Implement the following methods

#### (a) `add_fraction(other)`
This method should calculate the **sum of two fractions**.

Formula:

(a/b) + (c/d) = (ad + bc) / bd

---

#### (b) `subtract_fraction(other)`
This method should calculate the **difference of two fractions**.

Formula:

(a/b) - (c/d) = (ad - bc) / bd

---

#### (c) `multiply_fraction(other)`
This method should calculate the **product of two fractions**.

Formula:

(a/b) × (c/d) = (ac) / (bd)

---

#### (d) `divide_fraction(other)`
This method should calculate the **division of two fractions**.

Formula:

(a/b) ÷ (c/d) = (a/b) × (d/c)

---

#### (e) `simplify()`
This method should simplify the fraction to its **lowest terms** using the **Greatest Common Divisor (GCD)**.

Example:

6/8 → 3/4

---

#### (f) `display()`
This method should display the fraction in the form:

a/b

---

## Driver Program

Write a driver program to:

1. Create two fractions  
   - Fraction 1 = 3/4  
   - Fraction 2 = 5/6  

2. Perform the following operations:
   - Addition
   - Subtraction
   - Multiplication
   - Division

3. Display the results in **simplified fractional form**.

---

## Expected Output (Sample)

Fraction 1 : 3/4  
Fraction 2 : 5/6  

Addition : 19/12  
Subtraction : -1/12  
Multiplication : 5/8  
Division : 9/10

---

## Additional Instruction

- Do **not convert fractions to decimals**.  
- All calculations and outputs must remain in **fraction form only**.
- Use **objects of the class** to perform the operations.

```python
import math
class Fraction:
    def __init__(self, num, den):
        self.num=num
        self.den=den
    def __str__(self):
        return f"{self.num}/{self.den}"
    def add_fractions(self, other):
        num=self.num*other.den + self.den*other.num
        den=self.den*other.den
        return Fraction(num, den)
    def sub_fractions(self, other):
        num=self.num*other.den - self.den*other.num
        den=self.den*other.den
        return Fraction(num, den)
    def mul_fractions(self, other):
        num=self.num*other.num
        den=self.den*other.den
        return Fraction(num, den)
    def div_fractions(self, other):
        num=self.num*other.den
        den=self.den*other.num
        return Fraction(num, den)
    def simplify(self):
        g=math.gcd(self.num, self.den)
        return Fraction(self.num//g, self.den//g)

fraction=Fraction(2,3)
otherfraction=Fraction(5,6)
add=fraction.add_fractions(otherfraction)
sub=fraction.sub_fractions(otherfraction)
mul=fraction.mul_fractions(otherfraction)
div=fraction.div_fractions(otherfraction)

print(f"Fraction 1: {fraction}")
print(f"Fraction 2: {otherfraction}")
print(f"Addition result: {add.simplify()}")
print(f"Subtraction result: {sub.simplify()}")
print(f"Multiplication result: {mul.simplify()}")
print(f"Division result: {div.simplify()}")
```
