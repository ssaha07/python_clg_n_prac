# Question 1

A software company wants to develop a **Library Management System** using Object-Oriented Programming concepts.

### (a) Create a base class `Book` with the following private attributes:

- `__book_id`
- `__title`
- `__author`
- `__price`

Use a **class variable** `__book_counter` to automatically generate unique book IDs.

Implement the following methods:

1. `get_book_id()`  
   Returns the book ID.

2. `display_book_details()`  
   Displays book information.

---

### (b) Create a derived class `EBook` that inherits from the `Book` class.

Additional attributes:

- `file_size` (in MB)
- `download_link`

Override the `display_book_details()` method to display both **book details and ebook details**.

---

### (c) Create another derived class `PrintedBook` that inherits from the `Book` class.

Additional attributes:

- `weight` (in grams)
- `shipping_cost`

Add a method:

`calculate_total_price()`

Total Price = `price + shipping_cost`

---

### (d) Driver Program

Write a program to perform the following:

1. Create an object of `EBook`
2. Create an object of `PrintedBook`
3. Display details of both books
4. Calculate the total price of the printed book.

---

# Expected Sample Output

EBook Details  
Book ID: 1  
Title: Python Programming  
Author: John Smith  
Price: 500  
File Size: 15 MB  
Download Link: www.books.com/python  

Printed Book Details  
Book ID: 2  
Title: Data Structures  
Author: Mark Lee  
Price: 650  
Weight: 450 grams  
Shipping Cost: 50  

Total Price: 700

```python
class Book:
    __book_counter=0
    def __init__(self,title, author, price):
        Book.__book_counter += 1
        self.__book_id=Book.__book_counter
        self.__title=title
        self.__author=author
        self.__price=price
    def get_book_id(self):
        return self.__book_id
    def get_title(self):
        return self.__title
    def get_author(self):
        return self.__author
    def get_price(self):
        return self.__price
    def display_book_details(self):
        return f"""
                _____________________
                       DETAILS
                _____________________
                BOOK ID : {self.__book_id}
                TITLE   : {self.__title}
                AUTHOR  : {self.__author}
                PRICE   : {self.__price}
                """
class EBook(Book):
    def __init__(self, title, author, price, file_size, download_link):
        super().__init__(title, author, price)
        self.file_size=file_size
        self.download_link=download_link
    def display_book_details(self):
        return f"""
                _____________________
                    EBOOK DETAILS
                _____________________
                BOOK ID       : {super().get_book_id()}
                TITLE         : {super().get_title()}
                AUTHOR        : {super().get_author()}
                PRICE         : {super().get_price()}
                FILE SIZE     : {self.file_size}
                DOWNLOAD LINK : {self.download_link}
                """
class PrintedBook(Book):
    def __init__(self, title, author, price, weight, shipping):
        super().__init__(title, author, price)
        self.weight=weight
        self.shipping=shipping
    def calculate_total_price(self):
        total=super().get_price()+self.shipping
        return total
    def display_book_details(self):
        return f"""
                ____________________
                PRINTED BOOK DETAILS
                ____________________
                BOOK ID       : {super().get_book_id()}
                TITLE         : {super().get_title()}
                AUTHOR        : {super().get_author()}
                PRICE         : {super().get_price()}
                WEIGHT        : {self.weight}
                SHIPPING COST : {self.shipping}

                
                TOTAL PRICE   : {self.calculate_total_price()}
                """

```

```python
ebook=EBook("Python Programming", "John Smith", 500, 15, "www.books.com/python")
print(ebook.display_book_details())
```

```python
print_book=PrintedBook("Data Structures", "Mark Lee", 650, 450, 50)
print(print_book.display_book_details())
```
