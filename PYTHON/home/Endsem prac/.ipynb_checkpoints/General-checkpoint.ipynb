```python
l=["Rebecca", "  this is    pretty", " hello kittie", "master  chef me"]
def single_space(s):
    c=0
    for i in range(len(s)):
        if not s[(i-1)].isspace() and s[i].isspace() and not s[(i+1)].isspace():
            c+=1
    return c
for i in l:
    print(f"{i}, has {single_space(i)} single space(s)")
```

```python
l1=[
    {"Name":"A", "Physics_marks":80, "Chemistry_marks":89, "Biology_marks":79},
    {"Name":"B", "Physics_marks":85, "Chemistry_marks":95, "Biology_marks":80},
    {"Name":"C", "Physics_marks":60, "Chemistry_marks":59, "Biology_marks":89}
]
name=list(map(lambda x:x["Name"], filter(lambda x :((x["Physics_marks"]+x["Chemistry_marks"]+x["Biology_marks"])/3)>80, l1)))
for i in name:
    print(f"{i} scored more than 80 %")
```

```python
d={}
t=tuple()
n=int(input("Number of inputs:"))
while(n>0):
    name=input("Enter name: ")
    marks1=int(input("Enter marks of sub1: "))
    marks2=int(input("Enter marks of sub2: "))
    marks3=int(input("Enter marks of sub3: "))
    t=(marks1,marks2,marks3)
    d[name]=t
    n=n-1
print(d)
tot={name:sum(t) for name, t in d.items()}
all_tot=list(tot.values())
for k, v in d.items():
    if all_tot.count(tot[k])==1:
        print(f"{k}, {v}")
    
```

```python
def swimmer():
    l=[]
    for i in range(4):
        d={}
        name=input("Enter name:")
        lap1_time=int(input("Enter lap1 time:"))
        lap2_time=int(input("Enter lap2 time:"))
        lap3_time=int(input("Enter lap3 time:"))
        d["Name"]=name
        d["Lap1_time"]=lap1_time
        d["Lap2_time"]=lap2_time
        d["Lap3_time"]=lap3_time
        l.append(d)
    return l
swimmer_lessthan_10=list(filter(lambda x:(x["Lap1_time"]<10 
                                          and x["Lap2_time"]<10 
                                          and x["Lap3_time"]<10)
                               , swimmer())
                        )
for i in range(len(swimmer_lessthan_10)):
    print(swimmer_lessthan_10[i])
```

```python
num=121
s=0
n=num
while(n>0):
    r=n % 10
    s=s*10+r
    n=n//10
print(s)
if(num==s):
    print("Is palindrome")
else:
    print("Not palindrome")
```

```python
string="ant apple bat bottle"
words=string.split(' ')
d={}
for word in words:
    key=word[0]
    if key not in d:
        d[key]=[word]
    else:
        d[key].append(word)
print(d)
```

```python
n=2
l=[]
for i in range(n):
    d={}
    name=input("Enter name= ")
    height=float(input("Enter height= "))
    weight=int(input("Enter weight= "))
    d['Name']=name
    d['Height']=height
    d['Weight']=weight
    d['BMI']=weight/(height*height)
    l.append(d)
    
print(l)
```

```python
s='banana'
alpha="abcdefghijklmnopqrstuvwxyz"
k=3
res=''
for i in s:
    if i.isalpha():
        pos=alpha.index(i)
        new_pos=(pos+k)%26
        res+=alpha[new_pos]
    else:
        res+=i
print(res)
```
