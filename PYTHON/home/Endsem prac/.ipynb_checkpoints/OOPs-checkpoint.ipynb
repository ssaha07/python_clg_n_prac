```python
class BankAccount:
    def __init__(self, accholder, balance):
        self.accholder=accholder
        self.balance=balance
    def deposit(self, depositamt):
        self.balance=self.balance + depositamt
        print(f"Rs. {depositamt} is deposited.")
    def withdraw(self, withdrawamt):
        if ((self.balance-withdrawamt) < 0):
            print(f"Insufficient balance, cannot withdraw")
        else:
            self.balance=self.balance-withdrawamt
            print(f"Rs. {withdrawamt} is withdrawn.")
    def check_balance(self):
        print(f"Available balance is:{self.balance}") 
    def display(self):
        return f"""
        _____________ACCOUNT DETAILS____________
        Account holder:  {self.accholder}
        Balance       :  {self.balance}
        _________________________________________"""
```

```python
name=input("Enter name:")
rec1=BankAccount(name, 1000)
depositamt=int(input("Enter amount to deposit:"))
rec1.deposit(depositamt)
withdrawamt=int(input("Enter amount to withdraw:"))
rec1.withdraw(withdrawamt)
print(rec1.display())
```

```python
class Employee:
    def __init__(self, empid, name, department, category):
        self.empid=empid
        self.name=name
        self.department=department
        self.category=category
    def calculate_salary(self):
        return 0
    def display(self):
        print(f"""
        ___________EMPLOYEE DETAILS__________
        EMPID      : {self.empid}
        NAME       : {self.name}
        DEPARTMENT : {self.department}
        TYPE       : {self.category}
        SALARY     : {self.calculate_salary()}
        ______________________________________""")
class FullTimeEmployee(Employee):
    def __init__(self, empid, name, department, category, monthly_salary):
        super().__init__(empid, name, department, category)
        self.monthly_salary=monthly_salary
    def calculate_salary(self):
        return self.monthly_salary * 1.15
class PartTimeEmployee(Employee):
    def __init__(self, empid, name, department, category, hourly_rate, hours_worked):
        super().__init__(empid, name, department, category)
        self.hourly_rate=hourly_rate
        self.hours_worked=hours_worked
    def calculate_salary(self):
        tot=self.hourly_rate * self.hours_worked
        if(self.hours_worked < 80):
            tot-=(0.10 * tot)
        else:
            tot=tot
        return tot
employees=[]
e1=FullTimeEmployee("E01", "Priya Sen", "Engineering", "Full-Time", 57500)
employees.append(e1)
e2=PartTimeEmployee("E02", "Arjun Das", "Marketing", "Part-Time", 150, 75)
employees.append(e2)
for i in employees:
    i.display()
highest_earner=max(employees, key=lambda x:x.calculate_salary())
print(f"Highest Earner: {highest_earner.name} -- Rs. {highest_earner.calculate_salary()}")
```

```python
class Student:
    def __init__(self, roll, name, m1, m2, m3, m4, m5):
        self.roll=roll
        self.name=name
        self.m1=m1
        self.m2=m2
        self.m3=m3
        self.m4=m4
        self.m5=m5
    
    def get_name(self):
        return self.name 
    
    def get_total(self):
        return self.m1+self.m2+self.m3+self.m4+self.m5
    def get_average(self):
        tot=self.get_total()
        return tot/5
    def get_grade(self):
        avg=self.get_average()
        grade=""
        if avg >= 80:
            grade="A"
        elif avg >= 60 and avg < 80:
            grade="B"
        elif avg >= 40 and avg < 60:
            grade="C"
        else:
            grade="F"
        return grade
    def is_pass(self):
        if (self.m1>=35 and self.m2>=35 and self.m3>=35 and self.m4>=35 and self.m5>=35):
            return "Pass"
        else:
            return "Fail"
    def __str__(self):
        return f"Roll: {self.roll} | Name: {self.name} | Total: {self.get_total()} | Average: {self.get_average()} | Status:{self.is_pass()}"
class ResultBoard:
    def __init__(self):
        self.result=[]
    def add_student(self, student):
        self.result.append(student)
        print(f"Added {student}")
    def get_toppers(self):
        toppers=list(map(lambda x: x.get_name(), 
                         filter(lambda x: x.get_average()>=80, self.result)))
        print(f"TOPPERS (AVG >= 80) : {[i for i in toppers]}")
    def get_failed(self):
        failed=list(map(lambda x: x.get_name(),
                       filter(lambda x: x.is_pass()=="Fail", self.result)))
        print(f"FAILED STUDENTS : {[i for i in failed]}")
s1=Student(1, "A", 30, 20, 25, 19, 20)
s2=Student(2, "B", 67, 78, 88, 99, 90)
s3=Student(3, "C", 30, 67, 87, 89, 91)
s4=Student(4, "D", 76, 98, 54, 61, 86)

res=ResultBoard()
res.add_student(s1)
res.add_student(s2)
res.add_student(s3)
res.add_student(s4)

res.get_toppers()
res.get_failed()
```
