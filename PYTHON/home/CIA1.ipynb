```python
""" 
WAPP to count the number of single-space within a given string. The string may contain multiple consecutive spaces such as double spaces
triple spaces or more. Only spaces that occurs individually, not part of consecutive spaces, should be counted
"""
#string=input("Enter string (containing multiple consecutive spaces such as double spaces triple spaces or more)=")
string="Hello  new world    ."
c=0
#w=string.split()
for i in range(len(string)):
    #print(f"{i}:{string[i]}")
    if not string[(i-1)].isspace() and string[i].isspace() and not string[(i+1)].isspace():
        c = c + 1
print(f"no. of individually occuring space: {c}")
```

```python
l=[{"Name":"AB", "phy_marks":60, "chem_marks":70, "bio_marks":80}, 
   {"Name":"BC", "phy_marks":70, "chem_marks":80, "bio_marks":90},
   {"Name":"CD", "phy_marks":80, "chem_marks":90, "bio_marks":80}
  ]
for i in range(len(l)):
    details_gt_80=list(map(lambda x: x["Name"], filter(lambda x: (x["phy_marks"] + x["chem_marks"] + x["bio_marks"])/3 >=80, l)))
print(details_gt_80)
```
