```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data={
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Exam_score": [35, 40, 50, 55, 65, 70, 78, 85]
}
df=pd.DataFrame(data)
df
```

```python
x=df['Hours_Studied']
y=df['Exam_score']
plt.scatter(x, y)
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_score')
plt.show()
```

```python
X=df[['Hours_Studied']]
Y=df['Exam_score']
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
hours=5
new_data=pd.DataFrame({"Hours_Studied":[hours]})
score=model.predict(new_data)
print(f"If a student studied for {hours} hours, his predicted marks is  {score[0]:<.2f}")
```

```python
predictions=model.predict(X)
error=mean_absolute_error(y, predictions)
print(f"Mean Absolute Error: {error:<.2f}")
```

```python
plt.scatter(X, Y, label="Actual Data")
plt.plot(X, predictions, label="Regression Line", color="red")
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_score')
plt.legend
plt.show()
```

```python
data2={
    "Years_Experience": [1,2,3,4,5,6,7],
    "Salary_LPA(lakhs)": [6, 10, 12, 17, 23, 30, 40]
}
df2=pd.DataFrame(data2)
df2
```

```python
X=df2[['Years_Experience']]
Y=df2['Salary_LPA(lakhs)']
model2 = LinearRegression()
model2.fit(X,Y)
```

```python
years=10
new_data=pd.DataFrame({"Years_Experience":[years]})
sal=model2.predict(new_data)
print(f"If an employee worked for {years} years, his predicted salary is  {sal[0]:<.2f} lakhs")
```

```python
plt.scatter(X,Y, label="Actual Salary")
plt.plot(X,model2.predict(X), color="green", label="Predicted Salary")
plt.legend()
plt.xlabel("Years_Experience")
plt.ylabel("Salary")
plt.show()
```
