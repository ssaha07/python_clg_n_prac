```python
import numpy as np
import pandas as pd
```

```python
import math
marks=pd.Series([54,67,87,88,45,72], index=['English', 'Maths', 'Science', 'Art', 'History', 'Computer'])
print(marks.apply(lambda x:f"{math.sqrt(x):<.2f}"))
```

```python
def grades(marks):
    grade=""
    if marks >= 80:
        grade="A"
    elif marks >=60 and marks <=79:
        grade="B"
    else:
        grade="C"
    return grade
print(marks.apply(grades))
```

```python
s=pd.Series([20,21,12], index=['London', 'New York', 'Helsinki'])
def add_custom_values(x, **kwargs):
    for i in kwargs:
        x+=kwargs[i]
    return x
print(s.apply(add_custom_values, june=30, july=20, august=25))
```

```python
s=pd.Series([1,2,3,4,5,6])
def add_custom_values(x, a, b):
    x=x-a+b
    return x
print(s.apply(add_custom_values, args=(5,6)))
```

```python
s=pd.Series([1,2,6.1,856.87,'hello','Bye', 10, 'Balloon', '67.54'])
print(s.apply(lambda x: np.nan if str(x).isalpha() else x))
```

```python
print(s.apply(lambda x: float(x) if str(x).replace('.','', 1).isdigit() else x))
```

```python
random=np.random.randint(1, 100, size=40)
s=pd.Series(random)
chigh=s[s>=50].count()
clow=s[s<50].count()
print(chigh, clow)
def classify(s):
        if s>=70:
            return 'High'
        elif s>=69 and s<=40:
            return 'Medium'
        else:
            return 'Low'
cls=s.apply(classify)
print(cls)
odd=list(filter(lambda x: x%2 != 0, s.values))
print(odd)
```

```python
ser=pd.Series([" alice ", "BOB", "ChArLie", " david", "EVE"])
def clean(ser):
    ser=ser.strip()
    ser=ser.title()
    return ser
ser1=ser.apply(clean)
print(ser1)
```

```python
ser1.drop_duplicates()
```

```python
l=[i for i in range(10, 21) if i % 2 == 0]
s=pd.Series(l)
print(s)
def sq(s):
    if s>14:
        return s**2
    else:
        return s
print(s.apply(sq))
print(s.sum(), s.mean(), s.median())
```
