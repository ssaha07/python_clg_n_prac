```python
def palindrome(n):
    s=0
    m=n
    while(n>0):
        dig=n%10
        s=s*10+dig
        n=n//10
    if(m==s):
        return m
n=int(input("Enter n: "))
l=[]
for i in range(n):
    x=int(input("Enter no.: "))
    l.append(x)
m=list(filter(palindrome, l))  
print(f"palindrome numbers= {m}")
```

```python
# 2/9 - 5/13 + 8/17 -....(7terms)
n=7
s=0
for i in range(1, n+1):
    s=s+(((-1)**(i-1))*((3*i-1)/(4*i+5)))
print(s)
```

```python
# 1^2 + 3^2 + .... + n^2
n=int(input("Enter n: "))
s=0
for i in range(1, 2*n, 2):
    s=s+(i**2)
print(s) 
```

```python
def factorial(n):
    f=1
    for i in range(1, n+1):
        f=f*i
    return f
n=int(input("Enter n:"))
s=0.0
for i in range(n):
    s=s+(1/factorial(i))
print(f"{s:<.2f}")
```

```python
x=int(input("Enter x: "))
n=int(input("Enter n: "))
s=0
for i in range(1, n+1):
    s=s+(((-1)**(i+1)) * ((x**i)/factorial(i)))
print(s)
```

```python
n=int(input("enter n: "))
mid=n//2
for i in range(mid+1):
    print(" " *(mid-i)+"*" *(2*i+1))
for i in range(mid-1, -1, -1):
    print(" " *(mid-i)+"*" *(2*i+1))
```

```python
string=input("Enter string: ")
print("----------------------")
print("1. no. of vowels")
print("2. no. of consonents")
print("3. no. of digits")
print("4. no. of special chars")
print("5. no. of whitespaces")
print("6. no. of uppercases")
print("----------------------")

nv=0
nc=0
nd=0
nsp=0
nwsp=0
nup=0
l=[]
while(1):
    ch=int(input("Enter choice"))
    match ch:
        case 1:
            for i in string:
                if i.isalpha():
                    v=i.lower()
                    if v in ('aeiou') and v not in l:
                        l.append(v)
            print(f"No. of vowels = {len(l)}")
        case 2:
            for i in string:
                if i.isalpha():
                    if i not in ('AEIOUaeiou'):
                        nc=nc+1
            print(f"No. of consonants = {nc}")
        case 3:
            for i in string:
                if i.isdigit():
                    nd=nd+1
            print(f"No. of digits = {nd}")
        case 4:
            for i in string:
                if (not i.isalnum() and not i.isspace()):
                    nsp=nsp+1
            print(f"No. of special characters = {nsp}")
        case 5:
            for i in string:
                if i.isspace():
                    nwsp=nwsp+1
            print(f"No. of whitespaces = {nwsp}")
        case 6:
            for i in string:
                if i.isupper():
                    nup=nup+1
            print(f"No. of uppercase = {nup}")
        case _:
            print("Wrong choice")
    x=input("do you want to continue? Y/N")
    if x in 'Nn':
        break
```
