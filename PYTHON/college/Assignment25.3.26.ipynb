```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("company_sales_data.csv")
data
```

```python
x=data['month_number']
y=data['total_profit']
plt.plot(x,y)
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.xticks(x,x)
plt.yticks(np.arange(100000, 500001, 100000))
plt.show()
```

```python
x=data['month_number']
y=data['total_profit']
plt.plot(x,y,'--ro',mfc='black', linewidth=3, label='Profit data of last year')
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.legend(loc='lower right')
plt.show()
```

```python
face_cream=data['facecream']
face_wash=data['facewash']
tooth_paste=data['toothpaste']
bath_soap=data['bathingsoap']
shampoo=data['shampoo']
moisturizer=data['moisturizer']
months=data['month_number']
plt.plot(months,face_cream,'-bo', label='Face Cream sales data')
plt.plot(months,face_wash, color='orange',linestyle='-', marker='o', label='Face Wash sales data')
plt.plot(months,tooth_paste,'-go', label='Toothpaste sales data')
plt.plot(months,bath_soap,'-ro', label='Bathing Soap sales data')
plt.plot(months,shampoo,'-mo', label='Shampoo sales data')
plt.plot(months,moisturizer,color='brown',linestyle='-', marker='o', label='moisturizer sales data')

plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.legend()
plt.show()
```

```python
x=data['month_number']
y=data['toothpaste']
plt.scatter(x, y, label="Tooth paste Sales data")
plt.legend()
plt.title("Tooth paste Sales data")
plt.xlabel("Month number")
plt.ylabel("Number of units Sold")
plt.xticks(x,x)
plt.grid(linestyle="--")
```

```python
x=data['month_number']
y1=data['facecream']
y2=data['facewash']
width=0.35
plt.bar(x-width/2,y1, width=width, label='Face Cream sales data')
plt.bar(x+width/2,y2, width=width, label='Face Wash sales data')
plt.legend()
plt.title("Facecream and facewash sales data")
plt.xlabel("Month Number")
plt.xticks(x,x)
plt.ylabel("Sales units in number")
plt.grid(linestyle="--")
```

```python
x=data['month_number']
y=data['bathingsoap']
plt.bar(x, y)
plt.grid(linestyle="--")
plt.title("Bathing soap sales data")
plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.xticks(x,x)
plt.show()
```

```python
facecream=data['facecream']
facewash=data['facewash']
toothpaste=data['toothpaste']
bathingsoap=data['bathingsoap']
shampoo=data['shampoo']
moisturizer=data['moisturizer']
labels=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
values=[sum(facecream), sum(facewash), sum(toothpaste), sum(bathingsoap), sum(shampoo), sum(moisturizer)]
plt.figure(figsize=(10,10))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
x=data['month_number']
y1=data['facewash']
y2=data['bathingsoap']
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(x, y1, '-ko',linewidth=2)
plt.ylabel("Sales units in number")
plt.title("Sales data of a facewash")
plt.subplot(2,1,2)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(x, y2, '-ro', linewidth=2)
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Sales data of a bathingsoap")
plt.show()
```
