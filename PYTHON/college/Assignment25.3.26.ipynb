```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("company_sales_data.csv")
data
```

```python
x=data['month_number']
y=data['total_profit']
plt.plot(x,y)
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.show()
```

```python
x=data['month_number']
y=data['total_profit']
plt.plot(x,y,'--ro',mfc='black', linewidth=3)
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.legend('Profit data of last year')
plt.show()
```

```python
face_cream=data['facecream']
face_wash=data['facewash']
tooth_paste=data['toothpaste']
bath_soap=data['bathingsoap']
shampoo=data['shampoo']
moisturizer=data['moisturizer']
months=data['month_number']
plt.plot(months,face_cream,'-ro', label='Face Cream')
plt.plot(months,face_wash,'-go', label='Face Wash')
plt.plot(months,tooth_paste,'-bo', label='Toothpaste')
plt.plot(months,bath_soap,'-mo', label='Bathing Soap')
plt.plot(months,shampoo,'-bo', label='Shampoo')
plt.plot(months,moisturizer,'-co', label='moisturizer')

plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.legend()
plt.show()
```
