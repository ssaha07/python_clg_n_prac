```python
fruits=['Apple', 'Banana', 'Orange', 'Cucumber']
pd.Series(fruits, index=[1,2,3,4])
```

```python
num=[21,14,54,43]
pd.Series(num, index=[2,3,4,5])
```

```python
#list of even nos b2n 10 to 20, convert to series
L=[]
for i in range(10,21):
    L.append(i)
m=list(filter(lambda x: x%2==0,L))
pd.Series(m)
```

```python
l=[1,'A',-5,5.3,'Ram']
print(l)
pd.Series(l)
```

CUSTOM INDEX

```python
marks=[54,56,76,67]
sub=['English','Maths','Science','Art']
pd.Series(marks, index=sub)
```

NAMING A SERIES

```python
marks=[54,56,76,67]
sub=['English','Maths','Science','Art']
pd.Series(marks, index=sub, name='Marksheet')
```

```python
num=[21,14,54,43]
pd.Series(num, index=['a','b','c','d'], name='Number', dtype=float)
```

```python
num=[2,2,2,2]
pd.Series(num, index=['a','b','c','d'], name='Number', dtype=float)
```
