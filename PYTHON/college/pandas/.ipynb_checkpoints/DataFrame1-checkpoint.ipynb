```python
"""
what is a series? 
Series is a 1D labeled array built on Numpy ndarray 
main components of series : indices, values (index-labels: customised indexing)
2D : DataFrame : when we merge two or more series we get a dataframe
each column is a series and each row is a series
table inside python is a dataframe
"""
```

```python
import numpy as np
import pandas as pd
```

```python
ser1 = pd.Series(['Sania', 'Simran', 'Rumela', 'Ritika'])
ser2 = pd.Series([80, 78, 96, 68])
df = pd.DataFrame([ser1, ser2])
df
```

```python
df=pd.DataFrame({'name':ser1, 'score':ser2})
df
```

```python
l=[
    ['Sania', 80],
    ['Simran', 78],
    ['Rumela', 96],
    ['Ritika', 68]
]
dfl=pd.DataFrame(l, columns=['name', 'score'],index=['a','b','c','d'])
dfl.loc['c':'a':-1,:]     #loc includes the last index
```

```python
dfl.iloc[0:3,:]
```

```python
l=[
    ['Sania', 80],
    ['Simran', 78],
    ['Rumela', 96],
    ['Ritika', 68]
]
dfl=pd.DataFrame(l)
dfl
```

```python
d={
    'name':['Sania','Simran','Rumela','Ritika'],
    'score':[80, 78, 96, 68]
}
dfd=pd.DataFrame(d)
dfd
```
