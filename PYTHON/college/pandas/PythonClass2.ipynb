```python
s1=pd.Series([1,2,3,4,5,6])
```

```python
def add_custom_values(x, **kwargs):
    for month in kwargs:
        x+=kwargs[month]
    return x
s1.apply(add_custom_values, june=30, july=20, august=25)
```
