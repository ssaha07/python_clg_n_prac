```python
import numpy as np
import pandas as pd
```

```python
d1={'name':['Ram','Shyam','Sita'],
   'scores':[90,80,70]}
df1=pd.DataFrame(d1)
df1
```

```python
d={'name':['B','M'],
  'scores':[70,85]}
df2=pd.DataFrame(d)
df2
```

```python
pd.concat([df1, df2])
```

```python
t=pd.concat([df1, df2], ignore_index=True)
t
```

```python
df3=pd.DataFrame({'age':[23,25,27,22,30]})
df3
```

```python
df=pd.concat([t,df3],axis=1)
df
```
