```python
def even_num(x):
    return (x % 2 == 0)
num = int(input("Enter number of elements: "))
l2 = []
for i in range(num):
    x = int(input("Enter number"))
    l2.append(x)
final = list(filter(even_num, l2))
print(final)
```

```python
def isnotPrime(x):
    flag = 1
    for i in range(2, x):
        if (x % i == 0):
            flag = 0
    if(flag == 0):
        return True
    else:
        return False
l6 = [1, 2, 3, 4, 5, 6]
final6 = list(filter(isnotPrime, l6))
print(final6)
```

```python
def anagram(s1, s2):
    c1 = list(s1)
    c2 = list(s2)
    if (len(c1) == len(c2) and sorted(c1) == sorted(c2)) :
        return True
    else:
        return False
print(anagram("listen", "silent"))
```
