```python
"""accept the marks of student as input, compute the grade using the following rules
>= 90      grade = O
>= 80, <90 grade = A
>= 70, <80 grade = B
>= 60, <70 grade = C
>= 50, <60 grade = D
< 50       grade = F
"""
name = input("Enter name:")
a = int(input("Enter total marks: "))
if (a >= 90):
    grade = 'O'
elif (a >= 80 and a < 90):
    grade = 'A'
elif (a >= 70 and a < 80):
    grade = 'B'
elif (a >= 60 and a < 70):
    grade = 'C'
elif (a >= 50 and a < 60):
    grade = 'D'
else:
    grade = 'F'
print("Name of student: ", name)
print("Total marks of student: ", a)
print("Grade of student: ", grade)
```

```python
"""
Calculate the electricity consumption charges :
1. Fixed charge = 100 for everyone
2. units       tariff rate
   <= 100          @5
   >100 <=300      @7
   >300 <=500      @9
   >500            @13
3. GST = 18% to both fixed charge and tariff
"""
fixed = 100
tariff = 0
unit = float(input("Enter the units of electricity used up: "))
if (unit <= 100):
    tariff = unit * 5
elif unit <= 300:
    tariff = (100 * 5) + (unit - 100) * 7
elif unit <= 500:
    tariff = (100 * 5) + (200 * 7) + (unit - 300) * 9
else:
    tariff = (100 * 5) + (200 * 7) + (200 * 9) + (unit - 500) * 13
gst = 0.18 * tariff + 0.18 * fixed
total = tariff + gst
print("The electricity consumption charges: ", total)
```
