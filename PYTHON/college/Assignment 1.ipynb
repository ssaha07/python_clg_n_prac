```python
def chk_pass(x):
    u=0
    l=0
    d=0
    sp=0
    rep={}
    prev=""
    c=1
    if (len(x)>=8):
        for i in range(len(x)):
            if x[i].isupper():
                u=u+1
            elif x[i].islower():
                l=l+1
            elif x[i].isdigit():
                d=d+1
            elif x[i] in '!@#$%^&*':
                sp=sp+1
            if x[i]==prev:
                c=c+1
                if c>3:
                    return "Weak"
            else:
                c=1
        if u>=1 and l>=1 and d>=1 and sp>=1:
            return "Strong"
        else:
            return "Moderate"
    return "Weak"
            
l=["Admin123!","weakpass","STRONGpass99@","12345678","User#2024Secure"]
for i in l:
    print(f"{i}: {chk_pass(i)}")
```

```python
ser4=pd.Series([1,2,3,4,5,9,11,13,15])
def prime_check(x):
    if x < 2:
        return False
    for i in range(2,x**0.5):
        if x%i==0:
            return False
    return True
prime=ser4[ser4.apply(prime_check)]
prime
```
