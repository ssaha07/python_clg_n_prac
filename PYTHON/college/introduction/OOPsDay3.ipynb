```python
class Parent:
    def __init__(self, x, y):
        print("Hello from Parent")
        self.__x = x
        self.y = y
    def get_x(self):
        print(self.__x)
class Child(Parent):
    def test_private(self):
        print(f"x:{self.__x}")
obj=Child(100, 20)
obj.test_private()
```

```python
class Parent:
    def __init__(self, x, y):
        print("Hello from Parent")
        self.__x = x
        self.y = y
    def call_method(self):
        print("Calling from parent!")
class Child(Parent):
    def call_method(self):
        super().call_method()             #super():only works for member-functions, cannot be used to access data members
        print("Calling from child!")
obj=Child(100, 20)
obj.call_method()
```

```python
#Multi-level Inheritance
class Grand:
    def __init__(self):
        print("Hello from Grand")
    def call_method(self):
        return "Hi1"
class Parent(Grand):
    def __init__(self):
        print("Hello from Parent")
    def call_method(self):
        return "Hi2"
class Child(Parent):
    def __init__(self):
        print("Hello from child!")
obj=Child()
obj.call_method()
```

```python
#Hierarchical Inheritance:
class Parent:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        print("Hello from the parent")
class FirstChild(Parent):
    def __init__(self,p,q):
        self.p=p
        self.q=q
        print("Hello from the 1stChild")
class SecondChild(Parent):    
    pass
obj=FirstChild(100,20)
obj1=SecondChild(10,50)
```

```python
#Abstraction
# @(name) : decorator
from abc import ABC , abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass
class Car(Vehicle):
    def start(self):
        print("Car: Turn key -> Engine starts")
    def stop(self):
        print("Car: Press brake -> Engine stops")
class Motorcycle(Vehicle):
    def start(self):
        print("Motocycle: Turn key -> Engine starts")
    def stop(self):
        print("Motocycle: Press brake -> Engine stops")
class Scooter(Vehicle):
    def start(self):
        print("Scooter: Turn key -> Engine starts")
    def stop(self):
        print("Scooter: Press brake -> Engine stops")
vehicles=[Car(),Motorcycle(),Scooter()]
print("Starting all vehicles: ")
for vehicle in vehicles:
    vehicle.start()
print("Stopping all vehicles: ")
for vehicle in vehicles:
    vehicle.stop()
```

```python
"""
Create a fraction class eg: 5/8, 15/6
attributes-numberator, denominator
methods-init, str
        1. add()
        2. sub()
        3. mul()
        4. div()
"""
import math
class Fraction:
    def __init__(self, num, deno):
        gcd=math.gcd(num, deno)
        self.num=num//gcd
        self.deno=deno//gcd
    def __str__(self):
        return f"{self.num}/{self.deno}"
    def add(self, other):
        num1=self.num*other.deno + other.num*self.deno
        den1=self.deno*other.deno
        return Fraction(num1, den1)
    def sub(self, other):
        num1=self.num*other.deno - other.num*self.deno
        den1=self.deno*other.deno
        return Fraction(num1, den1)
    def mul(self,other):
        num1=self.num*other.num
        den1=self.deno*other.deno
        return Fraction(num1, den1)
    def div(self, other):
        num1=self.num*other.deno
        den1=self.deno*other.num
        return Fraction(num1, den1)
    def simplify(self):
        gcd=math.gcd(self.num, self.deno)
        return Fraction(self.num//gcd, self.deno//gcd)
obj=Fraction(5,6)
obj1=Fraction(7,6)
add=obj.add(obj1)
sub=obj.sub(obj1)
mul=obj.mul(obj1)
div=obj.div(obj1)
print(f"Addition of {obj} and {obj1} = {add.simplify()}")
print(f"Difference between {obj} and {obj1} = {sub.simplify()}")
print(f"Product of {obj} and {obj1} = {mul.simplify()}")
print(f"Quotient after dividing {obj} by {obj1} = {div.simplify()}")
```
