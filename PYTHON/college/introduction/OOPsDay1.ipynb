```python
#Constructors : object creation calls constructors implicitly
#1. Parameterised constructor
class DemoClass:
    def __init__(self, x, y):          #magic method / dunder method     __init__ has setup/configuration related codes
        self.x=x
        self.y=y   
    def __str__(self):                 #magic method /  dunder method
        return f"{self.x}....{self.y}"
    def test_method(self):     
        print("Hello")
```

```python
obj=DemoClass(4, 5)
```

```python
import math
class Point:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"({self.x}, {self.y})"
    def distance_to(self, other_point):
        x1=other_point.x - self.x
        y1=other_point.y - self.y
        dis=math.sqrt((x1**2)+(y1**2))
        return dis
    def distance_to_origin(self):
        dis=math.sqrt((self.x**2) + (self.y**2))
        return dis
    #def lies_on_line(self, line):
        #if(line.x+line.y==0):
            #return True
       # return False
obj=Point(4,5)
print(obj)
```

```python
import math
class Line:
    def __init__(self, a, b, c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"Line: {self.a}x + {self.b}y + {self.c} = 0"
    def distance_from_point(self, x, y):
        num=abs((self.a * x) + (self.b * y) + (self.c))
        dem=math.sqrt((self.a ** 2) + (self.b ** 2))
        dis=num/dem
        return dis
    def distance_from_line(self, line):
        num=abs(self.c-line.c)
        dem=math.sqrt(self.a ** 2 + self.b ** 2)
        dis=num/dem
        return dis
    
objp=Line(5, 6, 8)
objq=Line(5, 6, -10)
print(objp)
print(objq)
print(f"Distance from point: {objp.distance_from_point(1, 2):<.2f}")
print(f"Distance from line: {objp.distance_from_line(objq):<.2f}")
```

```python
#complex number
class ComplexNumber:
    def __init__(self, a, b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}i + {self.b}"
    def __abs__(self):
        return math.sqrt(self.a ** 2 + self.b ** 2)
    def sum(self, other):
        a=other.a
        b=other.b
        sumx=a+self.a
        sumy=b+self.b
        return f"Sum of 2 complex numbers: {sumx}i + {sumy}"
obj=ComplexNumber(5,1)
print(f"Complex Number 1: {obj}")
print(f"Magnitde of 1: {obj.magnitude():<.2f}")
```
