```python
def min_3_num(a,b,c):
    l2=[a,b,c]
    small=min(l2)
    return small
print("Minimum of three numbers :",min_3_num(12,3,21))
```

```python
def fact(n):
    f=1
    for i in range(1,n+1):
       f=f*i
    return f
print("The factorial=",fact(6))
```

```python
def isprime(num):
    for i in range(2,num):
        if num%i==0 :
            return 'False'
            break
    return 'True'
print(isprime(23))
```

```python
def count_vowels(text):
    count=0
    for i in text:
        if i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i =='A' or i=='E' or i=='I' or i=='O' or i=='U':
           count=count+1
    return count
line=input("Enter a line of text : ")
print("Number of vowels in the line :",count_vowels(line))
```

```python
def primefactors(num2):
    l6=list()
    j=2
    while num2>0 and j<=num2:
        if num2%j==0 :
            l6.append(j)
            num2=num2/j
        else:
            j=j+1
    print(l6)
primefactors(24)
```

```python
def findvowel(text):
    low=text.lower()
    l5=[]
    for i in low:
        if i in 'aeiou':
            if i not in l5:
                l5.append(i)
    return l5
line=input("Enter a line of text : ")
print("List of vowels in the line :",findvowel(line))
```
