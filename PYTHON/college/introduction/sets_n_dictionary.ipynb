```python
"""
1. key:numbers, values:num to words"""
d = {}
n = int(input("How many inputs = "))
for i in range(n):
    p=int(input("Enter key (int): "))
    w=input("to words: ")
    d[p] = w
e = int(input("Enter element to look up: "))
for k, v in d.items():
    if(k == e):
        print(k, " : ", v)
```
