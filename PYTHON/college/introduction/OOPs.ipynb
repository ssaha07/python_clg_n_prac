```python
#constructors
class Student:
    institute = "st.xavier's"
    #default constructors
    def __init__(self):
        pass
    def __init__(self, fullName, marks):
        self.name=fullName
        self.marks=marks
    #static method
    def printWish():
        print("I wish I had a puppy")
    def printWish2(self):
        print(f"I am {self.name} and I want coffee.")
    def printName(self):
        Student.printWish()
        self.printWish2()
        print(f"name: {self.name} college : {Student.institute} and marks : {self.marks}")
s1=Student("Aditya", 85)
s2=Student("Riddhi", 90)
s1.printName()
s2.printName()
```

```python
'''Abstraction: Hiding the implementational details of a class and only showing the essential features to the user.
   Encapsulation: Wrapping data and functions into a single unit(object). Basically
   data method
   '''
class Car:
    def __init__(self):
        self.acc=False
        self.brake=False
        self.clutch=False
    def start(self):
        self.clutch=True
        self.acc=True
        print("Car Started")
    def stop(self):
        self.acc=False
        self.brake=True
        print("Car Stopped")
car1=Car()
car1.start()
car2=Car()
car2.start()
car2.stop()
car1.stop()
```

```python
#Inheritance:
class Employee:
    def __init__(self, name, id):
        self.name=name
        self.id=id
    def showDetails(self):
        print(f"The name of the employee {self.id} is {self.name}")
class Programmer(Employee):
    def show(self):
        print("Python")
e1=Employee("Archit", 101)
e1.showDetails()

p1=Programmer("Bipasha", 102)
p1.showDetails()
p1.show()
        
```

```python
class A:
    def feature1(self):
        print("This is feature 1 of class A")
    def feature2(self):
        print("This is feature 2 of class A")
class B(A): #single-level inheritance
    def feature3(self):
        print("This is feature 3 of class B")
    def feature4(self):
        print("This is feature 4 of class B")
class C(B): #multi-level inheritance
    def feature5(self):
        print("This is feature 5 of class C")
    def feature6(self):
        print("This is feature 3 of class C")
if __name__ == "__main__":
    a1=A()
    a1.feature1()
    
```

```python
class LearningStrategyDominant:
    def learning_technique(self):
        print("Backpropagation")
class LearningStrategyNew:
    def learning_technique_new(self):
        print("Forward-Forward Algorithm")
class myLearningStrategyNew(LearningStrategyDominant, LearningStrategyNew):
    def learning_technique_mine(self):
        print("Forward Cooperation Backward Learning")
p1=myLearningStrategyNew()
p1.learning_technique()
```

```python
#super(): it is a built-in function that returns a proxy object
class A:
    def __init__(self):
        print("Initialising A")
class C:
    def __init__(self):
        print("Initialising C")
class B(A, C):
    def __init__(self):
        super().__init__()
        print("Initialising B")
b=B()
```
