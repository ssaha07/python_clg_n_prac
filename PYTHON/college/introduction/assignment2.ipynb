```python
n1 = int(input("Enter no. of rows: "))
for i in range(1, n1 + 1):
    for j in range(n1 - i):
        print(" ", end = " ")
    for k in range(2*i - 1):
            print("*", end = " ")
    print()
```

```python
n2 = int(input("Enter no of rows: "))
for i in range(1, n2 + 1):
    for j in range(1, n2 + 1):
        if((i == 1 or i == n2) or (j == 1 or j == n2)):
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
n3 = int(input("Enter no of rows in pascal's triangle: "))
for i in range(0, n3 + 1):
    print(" " * (n3 -i), end = " ")
    s = 1
    
    for j in range(i + 1):
        print(s, end = " ")
        s = s * (i - j) // ( j + 1 )
    print()
```

```python
n4 = int(input("Enter no of rows: "))
for i in range(1, n4 + 1):
    a = 1
    for j in range(1, i + 1):
        print(a, end = " ")
        a += 2
    print()
```

```python
n = int(input("No of rows = "))
mid = n // 2
for i in range(mid + 1):
    print(" " * (mid - i) + "*" * (2 * i + 1))
for i in range(mid - 1, -1, -1):
    print(" " * (mid - i) + "*" * (2 * i + 1))
    
    
```
