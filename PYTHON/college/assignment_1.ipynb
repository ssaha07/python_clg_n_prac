```python
"""
2. count 
i. no of vowels
ii. no of consonants
iii. no of spaces
iv. no of special characters
"""
nv = 0
nc = 0
ns = 0
nsp = 0
string = input("Enter a sentence: ")
for i in string:
    if(i in 'AEIOUaeiou'):
        nv = nv + 1
    elif i.isalpha():
        nc = nc + 1
    elif i.isspace():
        ns = ns + 1
    else:
        nsp = nsp + 1
print("no. of vowels: ", nv)
print("no. of consonants: ", nc)
print("no. of spaces: ", ns)
print("no. of special characters: ", nsp)
```

```python
"""
3. check whether a given no is prime or not
"""

ch = 1
n = int(input("Enter a number: "))
for i in range(2, n):
    if(n % i == 0):
        ch = 0
if(ch == 1):
    print("prime")
else:
    print("not prime")
```
