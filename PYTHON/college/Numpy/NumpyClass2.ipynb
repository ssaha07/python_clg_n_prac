N-DIMENSIONAL ARRAYS

```python
import numpy as np
l=[
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6]
]
arr=np.array(l)
print(arr,"\n", type(arr), "\n", arr.dtype, "\n", arr.size, "\n", arr.shape)
```

```python
print(arr[2,2])
```

```python
t=(
    (1,2,3),
    (2,3,4),
    (4,5,6)
)
arrt=np.array(t)
print(arrt,"\n", type(arrt), "\n", arrt.dtype, "\n", arrt.size, "\n", arrt.shape)
```

```python
B=arr.reshape(6,2)
print(arr)
print(B, B.shape)
```

```python
multizeros=np.zeros((5,3), dtype=int)
print(multizeros)
```

```python
multiones=np.ones((3,5), dtype=int)
print(multiones)
```

```python
fullarr=np.full((3,5), np.inf)
print(fullarr)
```

```python
iden=np.eye(3,4, dtype=int)
print(iden)
```

```python
A=np.arange(1,11)
B=np.split(A,5)
As=np.array_split(A,4)
print(B)
print(B[0])
print(B[0][0])
print(As)
```

```python
arr=np.random.randint(1,100,10)
even=arr[arr%2==0]
add=np.sum(even)
mean=np.mean(even)
print(f"Original array: {arr}")
print(f"Even nos. : {even}")
print(f"sum of all elements: {add}")
print(f"mean of all elements: {mean:<.2f}")
```

```python
A=np.random.randint(100, size=(4,5))
print(A)
print(f"sum of all elements: {A.sum()}")
print(f"sum of column-wise elements: {A.sum(axis=0)}")
print(f"sum of row-wise elements: {A.sum(axis=1)}")
```

```python
A=np.random.randint(100, size=(2,3,5,6))
print(A)
print(f"sum when axis =0: {A.sum(axis=0)}")
print(f"sum when axis =1: {A.sum(axis=1)}")
print(f"sum when axis =2: {A.sum(axis=2)}")
print(f"sum when axis =3: {A.sum(axis=3)}")

print(f"sum when axis =-1: {A.sum(axis=-1)}")
print(f"max when axis =0: {A.max(axis=0)}")
print(f"min when axis =1: {A.min(axis=1)}")
```

```python
A=np.random.randint(10, size=(2,3))
B=np.random.randint(10, size=(2,3))
S=A+B
D=A-B
M=A*B
#at=A@B
print(A)
print(B)
print(f"sum={S}")
print(f"diff={D}")
print(f"product={M}")
#print(f"matrix multiplication={at}")
```

```python
A=np.random.randint(10, size=(2,3))
B=np.random.randint(10, size=(3,2))
at=A@B
print(f"matrix multiplication=\n{at}")
```

```python
A=np.random.randint(10, size=(2,3))
B=np.random.randint(10, size=(3,2))
at=A@B
E=np.matmul(A,B)
print(A)
print(B)
print(at)
print(E)
```

```python
l1=[
    [1,2],
    [2,3]
]
l2=[
    [2,3],
    [3,4]
]
C=np.matmul(l1,l2)
dot=np.dot(l1, l2)
# @ doesnt work with lists
print(f"matrix multiplication:\n{C}")
print(f"dot product:\n{dot}")
```

```python
#input 2 matrices from user (3x3) find sum and product
l1=[]
l2=[]
for i in range(9):
    x=int(input(f"enter element{i+1}:"))
    l1.append(x)
A1=np.array(l1)
B1=np.reshape(A1, (3,3))
print("-----1st Matrix----")
print(B1)
for i in range(9):
    x=int(input(f"enter element{i+1}:"))
    l2.append(x)
A2=np.array(l2)
B2=np.reshape(A2, (3,3))
print("-----2nd Matrix----")
print(B2)
add=B1+B2
mul=np.matmul(B1, B2)
print(f"Sum of 2 matrices:\n {add}")
print(f"Product of 2 matrices:\n {mul}")
```
