```python
zeroarrf=np.zeros(5)
zeroarr=np.zeros(5, np.int32)
zeroarr1=np.zeros(5, dtype=int)
print(zeroarrf, zeroarrf.dtype)
print(zeroarr, zeroarr.dtype)
print(zeroarr1, zeroarr1.dtype)
```

```python
onesarrf=np.ones(5)
onesarr=np.ones(5, np.int32)
onesarr1=np.ones(5, dtype=int)
print(onesarrf, onesarrf.dtype)
print(onesarr, onesarr.dtype)
print(onesarr1, onesarr1.dtype)
```

```python
#define an array A containing odd nos 1-10, B containing even nos 1-10
A=np.arange(1,10,2)
B=np.arange(2,11,2)
print(f"ODD NO. ARRAY: {A}")
print(f"EVEN NO. ARRAY: {B}")
```

```python


C=A+B
print(f"Addition of A and B: {C}")
```

```python

M=np.dot(A,B)
print(M)
```

```python
new=np.arange(1,11)
print(new)
print(f"sum: {new.sum(axis=0)}")
print(f"mean: {new.mean(axis=0)}")
print(f"max: {new.max(axis=0)}")
print(f"min: {new.min(axis=0)}")
```

```python
print(f"Average: {np.average(new)}")   # average is a function of numpy library but sum, max, min, mean are member methods of class ndarray
#unlike mean, average() can be used to calculate weighted average
w=[2,1,2,1,2,1,2,1,2,1]
print(f"Average using weights: {np.average(new, weights=w):<.2f}")
```

```python
A=np.arange(1,10,2)
print(A)
B=np.arange(2,11,2)
print(B)
print(f"Median of nos 1 to 10 odd nos: {np.median(A)}")
print(f"Median of nos 1 to 10 even nos: {np.median(B)}")
```

```python
print(f"Standard Deviation of nos 1 to 10 odd nos: {np.std(A):<.2f}")
print(f"Varience of nos 1 to 10 odd nos: {np.var(A)}")
print(f"Standard Deviation of nos 1 to 10 even nos: {np.std(B):<.2f}")
print(f"Varience of nos 1 to 10 even nos: {np.var(B)}")
```

```python
#np.sort():in-place sorting, no new object is created, by default quick_sort
unsorted=np.array([2,3,6,5,8,7,0])
sorted_arr=np.sort(unsorted, kind='heapsort')
print(f"Original array : {unsorted}")
print(f"Sorted array   : {sorted_arr}")
```

```python
#slicing : access elements based on index
#filtering : access elements based on values
slicing=np.arange(1,11)
Z=slicing[::6]
print(Z)
```
