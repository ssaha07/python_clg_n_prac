```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("company_sales_data.csv")
data
```

```python
data.columns = data.columns.str.strip()
```

```python
months=data['month_number']
```

```python
totprofit=data['total_profit']
plt.plot(months, totprofit)
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(months,months)
plt.yticks(np.arange(100000, 500001, 100000))
plt.show()
```

```python
sold_units=data['total_profit']
plt.plot(months, sold_units, color='red', marker='o', mfc='black', linestyle='--', linewidth=3, label='Profit data of last year')
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.title("Company sales data of last year")
#the question said to plot total profit but the title as depicted in the picture is total sales, so done accordingly
plt.legend(loc='lower right')
plt.xticks(months, months)
plt.yticks(np.arange(100000, 500001, 100000))
plt.show()
```

```python
facecream=data['facecream']
facewash=data['facewash']
toothpaste=data['toothpaste']
bathingsoap=data['bathingsoap']
shampoo=data['shampoo']
moisturizer=data['moisturizer']
```

```python
plt.figure(figsize=(10, 10))
plt.plot(months, facecream, color='blue', marker='o', linestyle='-', label='Face cream sales data', linewidth=3)
plt.plot(months, facewash, color='orange', marker='o', linestyle='-', label='Face wash sales data', linewidth=5, alpha=0.5)
plt.plot(months, toothpaste, color='green', marker='o', linestyle='-', label='Toothpaste sales data', linewidth=3)
plt.plot(months, bathingsoap, color='red', marker='o', linestyle='-', label='Bathing soap sales data', linewidth=3)
plt.plot(months, shampoo, color='purple', marker='o', linestyle='-', label='Shampoo sales data', linewidth=3)
plt.plot(months, moisturizer, color='brown', marker='*', linestyle=':', label='Moisturizer sales data')
#face wash and moisturizer columns have the same data, hence the orange and brown lines respectively are overlapping
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Sales data")
plt.legend(loc='upper left')
plt.xticks(months, months)
plt.show()
```

```python
plt.scatter(months, toothpaste, label="Tooth paste Sales data")
plt.legend(loc='upper left')
plt.title("Tooth paste Sales data")
plt.xlabel("Month number")
plt.ylabel("Number of units Sold")
plt.xticks(months, months)
plt.grid(linestyle="--")
```

```python
width=0.4
plt.bar(months-width/2,facecream, width=width, label='Face Cream sales data')
plt.bar(months+width/2,facewash, width=width, label='Face Wash sales data')
plt.legend(loc='upper left')
plt.title("Facecream and facewash sales data")
plt.xlabel("Month Number")
plt.xticks(months,months)
plt.ylabel("Sales units in number")
plt.grid(linestyle="--")
```

```python
plt.bar(months, bathingsoap)
plt.grid(linestyle="--")
plt.title("Bathing soap sales data")
plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.xticks(months, months)
plt.show()
```

```python
labels=['Face Cream', 'Face Wash', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
values=[sum(facecream), sum(facewash), sum(toothpaste), sum(bathingsoap), sum(shampoo), sum(moisturizer)]
plt.figure(figsize=(9,9))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(months, facewash, color='black', linestyle='-', marker='o', linewidth=2)
plt.ylabel("Sales units in number")
plt.title("Sales data of a facewash")
plt.subplot(2,1,2)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(months, bathingsoap, color='red', linestyle='-', marker='o', linewidth=2)
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Sales data of a bathingsoap")
plt.show()
```
