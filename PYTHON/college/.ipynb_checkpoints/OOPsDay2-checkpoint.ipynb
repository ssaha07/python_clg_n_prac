ORDER SYSTEM

```python
class OrderSystem:
    
    def __init__(self, order_id, customer_name):
        self.__order_id=order_id
        self.__customer_name=customer_name
        self.__cart_total=0
        self.__discount_percent=10
        self.__final_amount=0
        self.__paid=False
    def add_to_cart(self, amount):                   
        self.__cart_total=self.__cart_total+amount
        return f"Items added! Cart total Rs.{self.__cart_total}"
    def get_discount(self, percent):            #setter method
        self.__discount_percent=percent
    def apply_discount(self):
        discount_amount=self.__cart_total*(self.__discount_percent/100)
        self.__final_amount=self.__cart_total-discount_amount
        return f"Discount applied! You save Rs. {discount_amount}"
    def process_payment(self):
        if (self.__final_amount==0):
            self.__apply_discount()
        if not self.__paid:
            self.__paid=True
            return f"Payment of Rs.{self.__final_amount} is successful!"
        return "Already Paid!"
    def show_details(self):
        return f"""
                 ORDER DETAILS
        _________________________________

        ORDER ID:         {self.__order_id}
        CUSTOMER NAME:    {self.__customer_name}
        CART TOTAL:       {self.__cart_total}
        DISCOUNT PERCENT: {self.__discount_percent}
        FINAL AMOUNT:     {self.__final_amount}
        PAYMENT STATUS:   {"PAID" if self.__paid else "PENDING"}

        __________________________________
        
        """
```

```python
obj1=OrderSystem(101, "Albert")
print(obj1.add_to_cart(600))
obj1.get_discount(20)        #getter method
print(obj1.apply_discount())
print(obj1.process_payment())
print(obj1.show_details())
```

```python
obj1=OrderSystem(102, "Rose")
obj1.__discount_percent=100                # '__' makes an attribute private, this wont allow to modify the private variables, 
                                           #  a new variable instance gets created
                                           # class variables are : _Class __varname (private) : Name Mangling
                                           # but obj1.__discount_percent : creates a new, different variable of this name, not related to the class.
print(obj1.add_to_cart(600))
print(obj1.apply_discount())
print(obj1.process_payment())
print(obj1.show_details())
```

```python
records=[]
class OrderSystem:
    __order_counter=1
    def __init__(self, customer_name):
        self.__order_id=OrderSystem.__order_counter
        OrderSystem.__order_counter=OrderSystem.__order_counter+1
        self.__customer_name=customer_name
        self.__cart_total=0
        self.__discount_percent=10
        self.__final_amount=0
        self.__paid=False
    @staticmethod
    def get_order_id():
        return OrderSystem.__order_counter
    def add_to_cart(self, amount):                   
        self.__cart_total=self.__cart_total+amount
        return f"Items added! Cart total Rs.{self.__cart_total}"
    def get_discount(self, percent):            #setter method
        self.__discount_percent=percent
    def apply_discount(self):
        discount_amount=self.__cart_total*(self.__discount_percent/100)
        self.__final_amount=self.__cart_total-discount_amount
        return f"Discount applied! You save Rs. {discount_amount}"
    def process_payment(self):
        if (self.__final_amount==0):
            self.__apply_discount()
        if not self.__paid:
            self.__paid=True
            return f"Payment of Rs.{self.__final_amount} is successful!"
        return "Already Paid!"
    def show_details(self):
        return f"""
                 ORDER DETAILS
        _________________________________

        ORDER ID:         {self.__order_id}
        CUSTOMER NAME:    {self.__customer_name}
        CART TOTAL:       {self.__cart_total}
        DISCOUNT PERCENT: {self.__discount_percent}
        FINAL AMOUNT:     {self.__final_amount}
        PAYMENT STATUS:   {"PAID" if self.__paid else "PENDING"}

        __________________________________
        
        """
```

```python
obj1=OrderSystem("Albert")
print(obj1.add_to_cart(600))
obj1.get_order_id()
obj1.get_discount(20)        #getter method
print(obj1.apply_discount())
print(obj1.process_payment())
print(obj1.show_details())
records.append(obj1.show_details())
```

INHERITANCE

```python
class Parent:
    def __init__(self, x):
        self.x=x
        print(f"This is x: {self.x} from Parent")
    def method_parent(self):
        print("Hello from parent")
class Child(Parent):
    def __init__(self, y):
        self.y=y
        print(f"This is y: {self.y} from Child")
    def method_child(self):
        print("Hello from child")
obj=Child(10)
obj.method_parent()
obj.method_child()
```
