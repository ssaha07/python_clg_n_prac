```python
n1 = int(input("Enter num 1: "))
n2 = int(input("Enter num 2: "))
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
ch = int(input("Enter choice: "))
match ch:
    case 1:
        print("Sum of ", n1, " and ", n2, " is ", (n1 + n2))
    case 2:
        print("Difference between ", n1, " and ", n2, " is ", (n1 - n2))
    case 3:
        print("Product of ", n1, " and ", n2, " is ", (n1 * n2))
    case 4:
        print("Quotient after dividing ", n1, " and ", n2, " is ", (n1 / n2))
    case _ :
        print("Invalid choice")
```

```python
#sum of first n natural nos.
n = int(input("Enter no of nos for sum: "))
s = 0
for i in range(1, n + 1):
    s = s + i
print("Sum of first ", n, " natural nos. is ", s)
```

```python
#same prog using while
n = int(input("Enter no of nos for sum: "))
s = 0
i = 1
while(i <= n):
    s = s + i
    i = i + 1
print("Sum of first ", n, " natural nos. is ", s)
```
