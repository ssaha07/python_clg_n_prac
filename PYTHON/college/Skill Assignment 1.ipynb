```python
"""
1.	Assume you have some messy product names as – 
" iphone-14 Pro ",
"SAMSUNG galaxy S23",
"oneplus_11",
"Pixel--7 Pro",
" xiaomi 13 "
Write Python code to do the following  – 
•	Make a 1-D representation of the given data 
•	Remove extra spaces
•	Replace _ and - with space
•	Convert to Title Case
•	Remove duplicate spaces
•	Output standardized product names
Use apply as and when require.
"""
```

```python
l=[" iphone-14 Pro ", "SAMSUNG   galaxy S23", "oneplus_11", "Pixel--7 Pro", " xiaomi 13 "]
arr=pd.Series(l)
print(f"""Original data\n{arr}""")
def clean_product(name):
    name=name.replace('_', ' ')
    name=name.replace('-', ' ')
    name=name.strip()
    
    result=""
    for i in range(len(name)):
        if name[i]==' ' and name[i-1]== ' ':
            continue
        result+=name[i]
    name=result.title()
    return name            
x=arr.apply(clean_product)
print(x)
```

```python
"""
2.	Consider a data given as follows which represents password – 
    "Admin123!",
    "weakpass",
    "STRONGpass99@",
    "12345678",
    "User#2024Secure"
Write Python code  to check the following and return that the password is “Strong”, “Moderate” or “Weak”
•	At least 8 characters
•	At least 1 uppercase
•	At least 1 lowercase
•	At least 1 digit
•	At least 1 special character
•	No repeating character more than 3 times consecutively
"""
```

```python
def chk_pass(x):
    u=0
    l=0
    d=0
    sp=0
    rep={}
    prev=""
    c=1
    if (len(x)>=8):
        for i in range(len(x)):
            if x[i].isupper():
                u=u+1
            elif x[i].islower():
                l=l+1
            elif x[i].isdigit():
                d=d+1
            elif x[i] in '!@#$%^&*':
                sp=sp+1
            if x[i]==prev:
                c=c+1
                if c>3:
                    return "Weak"
            else:
                c=1
        if u>=1 and l>=1 and d>=1 and sp>=1:
            return "Strong"
        else:
            return "Moderate"
    return "Weak"
            
l=["Admin123!","weakpass","STRONGpass99@","12345678","User#2024Secure"]
for i in l:
    print(f"{i}: {chk_pass(i)}")
```

```python
"""
4.	Write python code to create a numpy array of  size 10x10 randomly 
generated in between 1 to 100. Then extract all elements has higher
value than 50 and replace all elements divisible by 3 with -1. Find the 
indices of all elements that are multiples of 7 and Finally extract a 3x3 
subarray from the center. 
"""
```

```python
arr=np.random.randint(1,100, size=(10, 10))
print(arr,"\n",arr.shape,"\n")
greater_than_50=arr[arr > 50]
arr[arr % 3 == 0]=-1

print("\nElements greater than 50\n", greater_than_50)
print("\nModified Array:\n", arr)
print("\nIndices of elements that are multiple of 7:\n", np.argwhere(arr%7 == 0))
print("\n3x3 subarray:\n",arr[3:6,3:6])
```

```python
"""
5.	Write a python code to multiply two 2-D array without any 
predefined function. """
```

```python
arr1_rows=int(input("Number of rows in 1st matrix: "))
arr1_cols=int(input("Number of columns in 1st matrix: "))
arr2_rows=int(input("Number of rows in 2nd matrix: "))
arr2_cols=int(input("Number of columns in 2nd matrix: "))
while True:
    if arr1_cols != arr2_rows:   
        print("No. of columns of 1st matrix must be equal to no. of rows of 2nd matrix")
        arr2_rows = int(input("Re-enter number of rows in 2nd matrix: "))  
    else:
        break
arr1=np.random.randint(20, size=(arr1_rows,arr1_cols))
arr2=np.random.randint(20, size=(arr2_rows,arr2_cols))
print(f"1st Array:\n{arr1}")
print(f"2nd Array:\n{arr2}")
#print(f"Multiplication Array:\n{arr1@arr2}")
print("Matrix multiplication\n")
arr_res=np.zeros((arr1_rows,arr2_cols), dtype=int)
m=len(arr1)
n=len(arr1[0])
p=len(arr2[0])
for i in range(m):
    for j in range(p):
        for k in range(n):
            arr_res[i][j]+=arr1[i][k] * arr2[k][j]
for i in arr_res:
    print(i)
```

```python
"""6.	Write a python program to extract the prime numbers from a
given series.
"""
```

```python
import math
def isprime(x):
    if x==2:
        return True
    if x<2:
        return False
    for i in range(2, int(math.sqrt(x)+1)):
        if x%i==0:
            return False
    return True
    
l=[1,2,3,4,5,6,17,19]
ser1=pd.Series(l)
prime=ser1.apply(isprime)
for i in range(len(l)):
    print(f"{l[i]}-> {prime[i]}")
```
