```python
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.plot(x1,y1,label='line1')
plt.plot(x2,y2,label='line2')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Graph')
plt.grid(axis='y')
plt.legend(title='legend')
plt.show()
```

```python
roll=np.array([0,1,2,3])
phy=np.array([83,88,91,80])
plt.subplot(2,2,1)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(roll,phy,'go',ms=6)
plt.title("PHYSICS")

chem=np.array([60,77,80,78])
plt.subplot(2,2,2)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(roll,chem,'r*',ms=6)
plt.title("CHEMISTRY")

mat=np.array([90,89,87,67])
plt.subplot(2,2,3)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(roll,chem,'mv',ms=6)
plt.title("MATHS")

bio=np.array([89,87,88,56])
plt.subplot(2,2,4)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(roll,chem,'c^',ms=6)
plt.title("BIOLOGY")
plt.suptitle("Marks Distribution")
plt.show()
```

```python
x=np.array([5,15,35,25,10])
y=np.array([100,110,120,90,80])
colors=np.array([0,20,40,60,80])
sizes=np.array([200,300,400,500,600])
plt.scatter(x,y,s=sizes,c=colors,cmap='summer')
plt.colorbar()
plt.show()
```

```python
x=np.array([5,5.2,9,9.3,9.4])
y=np.array([18,18,45,30.4,30])
colors=np.array([0,20,40,60,80])
sizes=np.array([200,300,400,500,600])
plt.scatter(x,y,s=sizes,c=colors,cmap='spring', alpha=0.6)
plt.colorbar()
plt.show()
```

```python
months=np.arange(1,13)
rain_2024=np.array([10,20,30,40,100,120,150,200,300,280,90,40])
rain_2025=np.array([20,30,30,45,110,127,148,210,320,280,270,10])
colors=np.array([0,10,20,30,40,60,100,15,25,35,5,45])
plt.scatter(months,rain_2024,label="Rainfall_2024",c=colors,cmap='spring')
plt.scatter(months,rain_2025,label="Rainfall_2025",c=colors,cmap='winter')
plt.xlabel('Months')
plt.ylabel('Rainfall')
plt.legend()
plt.colorbar()
plt.show()
```
