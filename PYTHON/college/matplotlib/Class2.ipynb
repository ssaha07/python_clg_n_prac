```python
def maximum(rainfall):
    l=[]
    maxi=max(rainfall)
    for i in rainfall:
        if maxi==i:
            l.append(0.2)
        else:
            l.append(0)
    return l
rainfall=np.array([10,12,15,90,40,100,200,250,300,100,50,20])
lab=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

plt.pie(rainfall,labels=lab,explode=maximum(rainfall))
plt.show()        
    
    
```

```python
data={'Cat A':10,'Cat B':20,'Cat C':15,'Cat D':25,}
series=pd.Series(data)
plt.bar(series.index,series.values) # pass index as x axis and values as y axis
plt.title("Customized bar chart with Matplotlib")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
plt.pie(series.values)
plt.show()
```
